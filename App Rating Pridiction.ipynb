```python
import pandas as pd
import numpy as np

 
def convert_size(size):
    if size == 'Varies with device' or pd.isnull(size):
        return np.nan
    size = size.strip()
    if size.endswith('M'):
        return float(size[:-1]) * 1000  # Convert MB to KB
    elif size.endswith('k'):
        return float(size[:-1])         # Already in KB
    else:
        return np.nan

# Apply the function to create a new column with cleaned size in KB
df1['Size'] = df['Size'].apply(convert_size)
df1
```

```python
import numpy as np
 
def clean_installs(val):
    return pd.to_numeric(val.replace(',', '').replace('+', ''), errors='coerce')  
df1.loc[: ,'Installs'] = df1['Installs'].apply(clean_installs)
print(df1)
```

```python
def clean_price(val):
    return pd.to_numeric(val.replace('$', ''), errors='coerce') 
 
df1.loc[:,'Price'] = df1['Price'].apply(clean_price)

 
print(df1)
```
